```python
# -*- coding: utf-8 -*-
# @Time    : 2019/5/2
# @Author  : Divinerhjf
# @Software: Jupyter Notebook

import requests
from bs4 import BeautifulSoup
from collections import OrderedDict
import time 

keyword = '科创版'
pages = 10

# 定义抓取函数
def get_one_page(url):
    try:
        headers = {
            'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/65.0.3325.162 Safari/537.36'
        }
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            return response.text
        return None
    except RequestException:
        return None


# 定义解析函数
def parse_one_page(html):
    
    soup = BeautifulSoup(html, 'html.parser')
    heads = soup.findAll('h2')
    
    # 循环遍历，返回结果
    Data = OrderedDict()
    for head in heads:
        # 标题
        titleinfo = head.find('a')
        title = titleinfo.get_text()
        # url
        url = titleinfo['href']
        # 其他
        otherinfo= head.find('span', {'class':'fgray_time'}).get_text()
        source, date, time = otherinfo.split()
        # 整合字典
        Data[title] = [date, source, url]
    return Data


# 定义主调用函数
def main(keyword, page):
    url = "https://search.sina.com.cn/?c=news&ie=utf-8&q=" + str(keyword) + "&col=&range=&source=&from=&country=&size=&time=&a=&page=" + str(page) + "&pf=0&ps=0&dpc=1"
    html = get_one_page(url)
    print(str('第'+ str(page+1) + '页').center(40, '='))
    Data = parse_one_page(html)
    for item in Data:
        print('\n'.join([item] + Data[item]))
        print('\n')
        

if __name__ == '__main__':
    for i in range(pages):
        main(keyword, page=i)
        time.sleep(1)
```
